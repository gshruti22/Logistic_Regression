# Normalization to make the values comaparable to each other

```python
def normalize(x):
    norm=(x - np.mean(x,axis = 0))/np.std(x,axis = 0)
    return norm
x=normalize(x)
x
```

```python
y=data.label
y
y=y.values                         #convering y into numpy array
y=y.reshape(y.shape[0],1)
y
```

# Splitting x into Training, Test and Validation tests in order:60% train, 20% validation & 20 % test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
x_train, x_val, y_train, y_val = train_test_split(x_train,y_train, test_size=0.2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
print(x_val.shape)
print(y_val.shape)
```

```python
# m- numer of training examples, n-number of features
m,n=x_train.shape
m,n
```

```python

x_train.shape,y_train.shape
```

```python
#Defining sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
#model: y_predicted= c + m1x1 + m2x2 + m3x3 +....
#       y_predicted= theta0 + X*theta

iteration=50000                                                       #number of iterations
lr=0.0001                                                              #learning rate
theta0=0                                                                
theta=np.zeros((n,1))
error=[]                                                               #storing cost function values
for i in range(iteration):
    y_train_out=sigmoid(np.dot(x_train,theta)+theta0)                            # y_pred=(m,1) matrix
    cost=-np.mean(y_train*(np.log(y_train_out)) + (1-y_train)*np.log(1-y_train_out))
    dtheta=(1/m)*np.dot(x_train.T,y_train_out-y_train)
    dtheat0=(1/m)*np.sum(y_train_out-y_train)
    theta=theta-lr*dtheta
    theta0=theta0-lr*dtheat0
    error.append(cost)
    if i%10000==0:
        print(i,cost)
```

```python
plt.plot(range(iteration),error)
plt.show()
```

# Bianry classification of data to find whether there is a crack in road or not
1 corresponds to crack, 0 corrsponds to non crack

```python
def classify_crack(y_train_out):
    y_out= []
    for i in y_train_out:
        if i>=0.5:
            y_out.append(1)
        else:
            y_out.append(0)
    return y_out
```

```python
y_train_out=classify_crack(y_train_out)
len(y_train_out)
```

# Defining Accuracy

```python
def accuracy(y_train,y_train_out):
    t=0
    for i in range(len(y_train)):
        if y_train_out[i]==y_train[i]:
            t=t+1
    accuracy=(t*100)/len(y_train)
    return accuracy
```

```python
print('Accuracy of training set:', accuracy(y_train,y_train_out))
```

```python

x_val.shape,y_val.shape
```

```python
y_val_out=sigmoid(np.dot(x_val,theta)+theta0)
y_val_out=classify_crack(y_val_out)
print('Accuracy of validation set', accuracy(y_val,y_val_out))
```

```python

x_test.shape,y_test.shape
```

```python
y_test_out=sigmoid(np.dot(x_test,theta)+theta0)
y_test_out=classify_crack(y_test_out)
print("Test set accuracy" , accuracy(y_test,y_test_out))
```

```python
def F1score(y_pred,y_pred_out):
    TP=0
    FP=0
    TN=0
    FN=0
    for i in range(len(y_pred)):
        if y_pred[i]==1 and y_pred_out[i]==1:
            TP+=1
        elif y_pred[i]==0 and y_pred_out[i]==0:
            TN+=1
        elif y_pred[i]==1 and y_pred_out[i]==0:
            FP+=1
        elif y_pred[i]==0 and y_pred_out[i]==1:
            FN+=1
    precision=TP/(TP+FP)
    recall=TP/(TP+FN)    
    F1_score=2*(precision*recall)/(precision+recall)
    return F1_score
```

```python
print("F1 Score of validation set is", F1score(y_val,y_val_out))
print("F1 Score of test set is", F1score(y_test,y_test_out))
```
